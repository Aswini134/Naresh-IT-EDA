** one hot encoder**
- one hot encoder name says at a  time one will on and other will off
- for example case status has two labels
    - Certified
    - Denied
- when you apply one hot encoder on case status , it creates two more extra column
    - case_status_Certified
    - case_status_Denied
|case_status|
|-|
|Certified|
|Denied|

```python
|case_status||case_status_Certified||case_status_Denied|
|-||-||-|
|Certified|
|Denied|
```

$Advantages$
- when you develop ML model it is very imp the columns should be independent each other
- so here case satus creating two xtra columns
- which are independent each other, which means the row values at a time only one column has 1
- which means 90 degrees phase shift
- which means perpendicular to each other, orthogonal each other

$Disadvantages$
- the disadvantage is  if a column has 100 unique labels , 100 new columns will be created
- Data set will become SPARSE, which means huge
- columns are more means dimensions are more
-  The processin time is more,i.e; the memory consumption is more
- ** curse of dimensionality**
