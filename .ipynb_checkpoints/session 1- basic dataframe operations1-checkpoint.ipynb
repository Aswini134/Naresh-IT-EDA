$Add New column$

```python
name=["Ramesh","suresh","satish"]
age=[30,35,40]
name,age
data=zip(name,age)
df=pd.DataFrame(data,columns=["Name", "Age"],index=["A","B","C"])
df
```

- if you want to add a new column
- df ['new column']
- you need to have a list which is having some elements
- that elements nee dto equal to no.of rows
- city_names=['hyd','bnglr','chennai']
-df['city']=city_names


```python
city_names=["hyd","banglr","chennai"]
df['city']=city_names
df
```

$update$ $the$ $existing$ $column$

- if you want to create a new column or update the old column
- both are same way

```python
df['Name']=["ash","dino","Rhino"]
df['Age']=['30',"5","1"]
df['city']=['India','USA',"USA"]
df
```

$drop$ $the$ $column$

- in order to drop the column we need to give 3 parameters we need to use drop method
- 1. whether you need to drop column or row you need to decide
- 2.mention the row/column name
- axis
   - axis=1 reference as column ; axis=0 reference as row
    - you want to create a new dataframe or you want to overwrite the existing dataframe
    - inplace=True

```python
df.drop("city",axis=1, inplace=True)
df
```

```python
df.drop("A",axis=0, inplace=True)
df
```

$save$  $the$  $data frame$

```python
df.to_csv("output.csv")
df.to_excel("output.xlsx")
```
